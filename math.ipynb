```python
inr = input("Enter the character: ").strip()

if len(inr) == 1:
    print("Invalid input")
elif inr.isalpha():
    print("It is an alphabet")
elif inr.isdigit():
    print("It is a digit")
else:
    print("It is a special character")
```
