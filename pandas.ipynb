<p style="font-family:Fantasy">
<code>read_csv()</code> function in pandas is used to read data from CSV files into a pandas Dataframe
</p>
<code>pd.read_csv(filepath_or_buffer, sep=' ,' , header='infer',  index_col=None, usecols=None, engine=None, skiprows=None, nrows=None)</code>
<p>filepath_or_buffer: Location of the csv file. It accepts any string path or URL of the file.
sep: It stands for separator, default is ', '.
header: It accepts int, a list of int, row numbers to use as the column names, and the start of the data. If no names are passed, i.e., header=None, then, it will display the first column as 0, the second as 1, and so on.
usecols: Retrieves only selected columns from the CSV file.
nrows: Number of rows to be displayed from the dataset.
index_col: If None, there are no index numbers displayed along with records.  
skiprows: Skips passed rows in the new data frame.</p>

<code>describe()</code>
<p style="font-family:Cursive">Pandas generates descriptive statistics of DataFrame columns which provides key metrics like mean, standard deviation, percentiles and more. </p>

<code>DataFrame.describe(percentiles=None, include=None, exclude=None)</code>
<p style="font-family:Cursive">percentiles: A list of numbers between 0 and 1 specifying which percentiles to return. The default is None which returns the 25th, 50th and 75th percentiles.
include: A list of data types to include in the summary like int, float, object for strings. Default is None means all numeric types are included.
exclude: A list of data types to exclude from the summary. Default is None means no types are excluded.</p>

```python
data.describe()
```

<li>count: Total number of non-null entries in each column.</li>
<li>mean: Average (mean) of the values in the column.</li>
<li>std: Standard deviation showing how spread out the values are</li>
<li>min: Minimum value in the column</li>
<li>25%: The 25th percentile (Q1) which means 25% of the data points are less than this value.</li>
<li>50%: Median value (50th percentile) where half the data points are below it.</li>
<li>75%: The 75th percentile (Q3) means 75% of the data points are below this value.</li>
<li>max: Maximum value in the column.</li>

```python
desc=data.describe(percentiles=[.20,.40,.60,.80],include=['object','float','int'])
desc
```

<li>count: Total number of non-null values</li>
<li>unique: Number of unique values in the column.</li>
<li>top: Most frequent value in the column.</li>
<li>freq: Frequency of the most common value.</li>

<code>df.rename(columns={...}, inplace=True)</code>
<p style="font-family:Cursive">renames specified columns in the dictionary.</p>

```python
import pandas as pd
df = pd.DataFrame({'A': [10, 20], 'B': [30, 40], 'C': [50, 60]})
df.rename(columns={'A': 'X', 'B': 'Y'}, inplace=True)
print(df)
```
